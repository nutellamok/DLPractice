### **Autograd**
automatic differentiation for all operations on Tensors

```python
x = torch.ones(2, 2)
print(x)

x.requires_grad=True
print(x)

x = torch.zeros(2,2, requires_grad=True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
z = y * y * 3
out = z.mean()
print(x.grad)

out.backward()

print(x.grad)
```

```python
print(x.requires_grad)
print((x ** 2).requires_grad)

with torch.no_grad():
    print((x ** 2).requires_grad)
```

```python
print(x.requires_grad)
y = x.detach()
print(y.requires_grad)
```
